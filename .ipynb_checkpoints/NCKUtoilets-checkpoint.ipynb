```python
## 加入套件和參數
import requests
import folium

relation_id = 12731746

lat, lon = 22.998925371930955, 120.21736604111638
overpass_url = "https://overpass-api.de/api/interpreter"
overpass_query = f"""
[out:json];
area({relation_id})->.searchArea;
(node["amenity"="toilets"](around:{radius},{lat},{lon});
);
out center;
"""
response = requests.post(overpass_url, data=overpass_query)

data = response.json()

if response.status_code == 200:
    data = response.json()
    print("成功取得 JSON 資料")
```

```python
m = folium.Map(location=[lat, lon], zoom_start=15)

for element in data['elements']:
    if 'lat' in element and 'lon' in element:
        tags = element.get('tags', {})
        
        # 用 HTML 格式組 popup 內容
        popup_html = f"<b>OSM ID:</b> {element['id']}<br>"
        for k, v in tags.items():
            popup_html += f"<b>{k}:</b> {v}<br>"

        # 設定顏色：無障礙廁所顯示綠色
        color = 'green' if tags.get('wheelchair') == 'yes' else 'blue'

        folium.Marker(
            location=[element['lat'], element['lon']],
            popup=folium.Popup(popup_html, max_width=300),
            icon=folium.Icon(color=color, icon='info-sign')
        ).add_to(m)
     
```
